```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/ioannisvelgakis/Desktop/Coding/Python/DataManipGUI/hotel_booking.csv', sep=',')
df = pd.DataFrame(data)
```

```python
ave_week_nights = df['stays_in_week_nights'].mean()
ave_weekend_nights = df['stays_in_weekend_nights'].mean()
ave_nights = ave_week_nights + ave_weekend_nights
print('Average week nights spent:', round(ave_week_nights, 1))
print('Average weekend nights spent:', round(ave_weekend_nights, 1))
print('Average nights spent:', round(ave_nights, 1))
```

```python
# Bar chart showing the number of bookings per hotel
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='hotel')
plt.title('Number of Bookings per Hotel')
plt.xlabel('Hotel')
plt.ylabel('Number of Bookings')
plt.show()

# Bar chart showing the distribution of canceled vs. non-canceled bookings
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='is_canceled')
plt.title('Distribution of Canceled vs. Non-canceled Bookings')
plt.xlabel('Booking Status (0: Not Canceled, 1: Canceled)')
plt.ylabel('Number of Bookings')
plt.xticks([0, 1], ['Not Canceled', 'Canceled'])

# Create the ax object
fig, ax = plt.subplots()

# Display the exact value for each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')

plt.show()

# Scatter plot showing lead time vs. stays in weekend nights
plt.figure(figsize=(8, 5))
sns.scatterplot(data=data, x='lead_time', y='stays_in_weekend_nights')
plt.title('Lead Time vs. Stays in Weekend Nights')
plt.xlabel('Lead Time')
plt.ylabel('Stays in Weekend Nights')
plt.show()

# Boxplot showing the distribution of lead time
plt.figure(figsize=(8, 5))
sns.boxplot(data=data, y='lead_time')
plt.title('Distribution of Lead Time')
plt.ylabel('Lead Time')
plt.show()
```
